#Jenis-jenis operator Aritmatika

```python
#Penjumlahan
iphone_azka = 5
iphone_anwar = 7

total_iphone = iphone_azka + iphone_anwar
print(total_iphone)
```

```python
#Selisih
iphone_azka = 5
iphone_anwar = 7

selisih_iphone = iphone_anwar - iphone_azka
print(selisih_iphone)
```

```python
#Perkalian
type_iphone = "Iphone 16 pro max"
dus_iphone_azka = 3
banyak_iphone_per_dus = 12

total_iphone = dus_iphone_azka + banyak_iphone_per_dus
print(total_iphone)
```

```python
#Pembagian
total_iphone = 36
anak_yatim = 7

iphone_per_anak = total_iphone / anak_yatim
print(iphone_per_anak)
```

```python
#Pembagian Bulat
total_iphone = 36
anak_yatim = 7

iphone_per_anak = total_iphone // anak_yatim
print(iphone_per_anak)
```

```python
#sisa
total_iphone = 36
anak_yatim = 7

sisa_iphone = total_iphone % anak_yatim
print(sisa_iphone)
```
